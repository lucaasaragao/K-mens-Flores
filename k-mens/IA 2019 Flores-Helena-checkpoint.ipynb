Importando classes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sb

%matplotlib inline
```

DateFrame e colunas 

```python
df = pd.read_csv('treinamento.csv')
```

```python
df.columns
```

```python
df
```

```python
df.describe()
```

```python
sb.pairplot(df)
```

Atribuindo a variavel "X" um novo array de np
e excluindo a coluna 'label'

```python
X = np.array(df.drop('label' ,axis = 1))
```

```python
X
```

Importando K-mens 


```python
from sklearn.cluster import KMeans
```

Atribuindo ao k-MENS o numero de cluster e o onde ele ira iniciar


```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

## Adicionando K-classes com o Kmens já treinado

```python
df['Kclasses'] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df, hue='label')
```

```python
sb.pairplot(df,'Kclasses')
```
